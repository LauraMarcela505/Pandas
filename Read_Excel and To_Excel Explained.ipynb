# Pandas: Read_Excel and To_Excel *Explained*
- read_excel: function in Pandas that is used for importing data from Excel files into a Pandas DataFrame
- to_excel: function in Pandas used to write or export a Pandas DataFrame to an Excel

## <u>Libraries</u>

```python
import pandas as pd
```

# Read_Excel

## <u>Importing Data into a DataFrame

```python
path = "C:\\Users\\laura\OneDrive\\Desktop\Python Sandbox for Projects\\DataFrames Pandas\\Python Pandas Read_Excel in Pandas\CompanyData.xlsx"
df = pd.read_excel(path)
```

```python
df
```

## <u>Explaining Different Parameters

```python
df_all_sheets = pd.read_excel(path, sheet_name=['Employees','Financials'])
```

```python
df_all_sheets['Employees']
df_all_sheets['Financials']
```

```python
df_employees = pd.read_excel(path, sheet_name='Employees')
df_financials = pd.read_excel(path, sheet_name='Financials')
```

```python
df_financials
```

### Parameter: header

```python
df_financials = pd.read_excel(path, sheet_name='Financials',header=3)
```

```python
df_financials
```

### Parameter: nrows

```python
df_financials = pd.read_excel(path, sheet_name='Financials',header=3, nrows=100)
```

```python
df_financials
```

### Parameter: usecols

```python
df_financials = pd.read_excel(path, sheet_name='Financials',header=3, usecols=['Account', 'Business Unit'])
```

```python
df_financials
```

### Parameter: index_col

```python
df_financials = pd.read_excel(path, sheet_name='Financials',header=3, index_col='Account')
```

```python
df_financials
```

# To_Excel

```python
df_employees.to_excel('Employees_Data.xlsx', sheet_name='Employees', index=False)
```

```python
with pd.ExcelWriter('All_Company_Data.xlsx') as writer:
    df_employees.to_excel(writer, sheet_name='Employees', index=False)
    df_financials.to_excel(writer, sheet_name='Financials', index=False)
```
